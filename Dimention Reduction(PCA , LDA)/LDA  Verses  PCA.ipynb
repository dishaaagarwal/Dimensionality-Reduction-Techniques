```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap

%matplotlib inline
```

```python
from sklearn import datasets

df=datasets.load_iris()

X=df.data
y=df.target

target_names = df.target_names
```

```python
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_std=scaler.fit_transform(X)
```

```python
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_std)
```

```python
for i in (np.unique(y)):
    plt.scatter(X_pca[y==i,0],X_pca[y==i,1],edgecolor='k',linewidth=2,alpha=0.8,label=target_names[i])
    
plt.legend()
plt.title('PCA')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.tight_layout()
plt.show()
```

```python
lda=LDA(n_components=2)
lda.fit(X_std,y)
```

```python
X_lda=lda.transform(X_std)
```

```python
for i in (np.unique(y)):
    plt.scatter(X_lda[y==i,0],X_lda[y==i,1],edgecolor='k',linewidth=2,alpha=0.8,label=target_names[i])
    
plt.legend()
plt.title('LDA')
plt.xlabel('LD1')
plt.ylabel('LD2')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,7))

plt.subplot(1,2,1)
for i in (np.unique(y)):
    plt.scatter(X_pca[y==i,0],X_pca[y==i,1],edgecolor='k',linewidth=2,alpha=0.8,label=target_names[i])
    
plt.legend()
plt.title('PCA')
plt.xlabel('PC1')
plt.ylabel('PC2')


plt.subplot(1,2,2)
for i in (np.unique(y)):
    plt.scatter(X_lda[y==i,0],X_lda[y==i,1],edgecolor='k',linewidth=2,alpha=0.8,label=target_names[i])
    
plt.legend()
plt.title('LDA')
plt.xlabel('LD1')
plt.ylabel('LD2')

plt.tight_layout()
plt.show()
```
